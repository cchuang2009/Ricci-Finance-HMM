## 1. System architecture

```mermaid
flowchart TD
    A[Prices and volume] --> B[Returns and dollar volume]
    B --> C[Rolling windows]
    C --> D[Correlation and financial distance]
    D --> E[Threshold or kNN graph]
    B --> F[Capital mass and edge transport]
    E --> G[Capital-adjusted graph]
    F --> G
    G --> H[Ollivier–Ricci curvature]
    H --> I[Graph statistics by frame]
    I --> J[Gaussian HMM]
    J --> K[State probabilities]
    I --> L[Frame comparison]
    K --> L
    L --> M[Market story]
    H --> N[Stable 3-D manifold]
    M --> N
```

The key conceptual separation is:

- **Ricci curvature** describes spatial network geometry;
- **rolling frames** create temporal observations;
- **HMM** summarizes hidden structural regimes;
- **story generation** translates measured changes into descriptive language.

## 4. Validate data and calculate log returns

For ticker \(i\), the log return is

$$r_i(t)=\log\left(\frac{P_i(t)}{P_i(t-1)}\right).$$

Log returns are additive through time and are more convenient than raw prices for correlation analysis.

Question
---
<b>Why do we use log return $\log r_i$ </b>
1. why not use price?
   `The price itself is non-stationary:`
    - continuously increases
    - has trend
    - variance changes with time <br>
`correlation` assumed approximately stationary!
2. Why use `log`? it makes data more normalized!
  - $5 \to 10$, increases 100%, $10 \to 5$ decreases   50%, un-resonable; but log * gives log 2 and -log2, same quantity but alternative signs.
  - log owns `additive` property in windows mode: 
$$\log\frac{P_{i+2}}{P_i} = \log\frac{P_{i+1}}{P_i}+ \log\frac{P_{i+2}}{P_{i+1}}$$

Ricci, and  HMM data hold also.

## 6. Correlation and financial distance

The correlation distance is


$$d_{ij}^{\mathrm{corr}}=\sqrt{2(1-\rho_{ij})}.$$


Interpretation:

- $\rho_{ij}\approx 1$: short distance;
- $\rho_{ij}\approx 0$: intermediate distance;
- $\rho_{ij}\approx -1$: long distance.

Question
---
Why use finacial distance to replace of Pearson correlation?

Answer: 

1. Fiancial distance is like `geomtric` distance, while 

$\rho = -1, 0 ,1 \to d= 2,\sqrt2, 0$ means very long, intermediate, same position.  
2.  Ollivier–Ricci curvature compares neighborhoods using optimal transport.        
$$ κ(x,y)=1−\frac{W(m_x,m_y)}{d(x,y)},$$
where 
 - $W$ is the Wasserstein distance,
 - $m_x,m_y$ are neighborhood probability distributions,
 - $d(x,y)$ is the edge length.

$d(x,y)$ can not be zero, can not be negative, (length of edge $\ge0$). If $$ is small, e.g. 0.3, means "These stocks are close in the market manifold."

 - Ricci flow:, update the Ricci iterated till convergent
$$ d_{ij}^0=d_{ij}, d_{ij}^{k+1}=d_{ij}^k\left( 1-\epsilon\kappa_{ij}^k  \right)$$
where $ \epsilon$ is a small scale; After convergence, the edge lengths encode both: *a)* the original return similarity, *b)* the surrounding network topology.

````python
WINDOW_SIZE = 60
window_returns = returns.iloc[-WINDOW_SIZE:]

corr = window_returns.corr(min_periods=20)
dist = financial_distance_from_corr(corr)

display(corr.round(3))
display(dist.round(3))
````

## 7. Build the financial graph

**Method:** each node is a ticker. Edges are selected by the chosen graph rule.

Here we use `knn+bridges`:

- k-nearest-neighbor edges preserve local relationships;
- a small number of additional bridge edges preserve important cross-cluster connections;
- overlap counts and confidence values are stored on each edge.

````python
G0, corr_frame, dist_frame = build_graph_from_window(
    window_returns,
    graph_mode="knn+bridges",
    k_neighbors=3,
    min_corr=0.05,
    max_bridges=3,
    min_pair_obs=20,
)

print("Nodes:", G0.number_of_nodes())
print("Edges:", G0.number_of_edges())
display(summarize_edges(G0))
````

## 8. Add capital mass and edge transport

Correlation tells
```
Do two stocks move together?
```
but not tells
```
Are these movements occurring in two highly traded stocks, or in two thinly traded stocks?
```


To enchance this invesgation, introduce, for each node,  median dollar volume in the current window as a capital-mass proxy.

The edge transport proxy is


$$F_{ij}=\sqrt{M_iM_j}\max(\rho_{ij},0).$$

where $M_i,M_j$	are the node capital masses, $\rho_{ij}$ is the profit correlation; negative correlation ($\rho <0$), no transport; highly correlated ticks is approximate their geometric mean.

Consider from statistical theory ($\rho$) to economical theory, define  
the effective distance:

$$d_{ij}^{\mathrm{eff}}=\frac{d_{ij}^{\mathrm{corr}}}{1+\alpha_c s_{ij}},$$

where $s_{ij}$ measures similarity in capital scale. Formally, $d_{ij}^{\mathrm{eff}}\le d_{ij}^{\mathrm{corr}}$

### Why both are needed

- Using only the transport proxy would lose a clean metric interpretation.
- Using only correlation distance would ignore the economic scale of trading activity.

### Together, 
they separate:

- edge length: how far apart the stocks are;
- edge capacity or activity: how economically important the relationship is.  

# Logic Analogy:

- effective distance = how difficult or long the route is;
- transport proxy = how much traffic can plausibly use the route;
- After while, introduce `Ricci curvature`,  whether the route lies inside a dense network or acts as a fragile bridge.

### Formula
1. Daily capical volume, $DV_i(t)= P_i(t)V_i(t)$, the deal amount of tick $i$ at time $t$, 
2. node capital mass, calculated over the rolling window:
   $$ M_i = median_{t\in Windows} DV_i(t)$$
3. similarity, $0\le Ss{ij}\le1$:
   $$ s_{ij} = 1-\frac{|M_i-M_j|}{M_i+M_j}$$    

| $s_{ij}$| Interpretation|
|---|---|
|1|Equal capital masses|
|0.9|Very similar liquidity scales|
|0.5|different scales|
|0|One node is much larger than the other|

This is a modeling proxy—not direct evidence that cash physically moved from one stock to another.

````python
capital_window = dollar_volume.reindex(
    index=window_returns.index,
    columns=window_returns.columns,
)

G_capital = attach_capital_attributes(
    G0,
    capital_window,
    capital_alpha=0.35,
    use_capital_weighting=True,
)

display(node_capital_table(G_capital))
display(capital_flow_table(G_capital).head(10))
````

## 9. Compute Ollivier–Ricci curvature

### Why introduce Ricci curvature

```                     / D
A --- B ---C,    A --- B  --- C
                        \ E
```
the correlation of A and B in both charts are identitcal, but diffewrentical geometrically. use Ricci cruvature to distinct them.

### Formula
$$ \kappa_0(i,j) = 1-\frac{\text{Waserstasin's distance}(i,j)}{d_0{i,j}}$$

where 
$$ W(\mu,\nu)= \inf_{\pi\in\Pi(\mu,\nu)}\sum_{(x,y)\in X}d(x,y)\pi(x,y)$$

### Example
```
Case1.

Nvdia --- Broadcom --- AMD

neighborhood of N: m1 = (0.5,0.5,0)
nbd of AMD: (0,0.5,0.5)
|
For B, only $A\to B$, $0.5\times1$, $C\to B$, $0.5\times1$, theirfore $W(m_1,m_2)=0.5$

Case2.

Nvdia --- Broadcom --- AMD

neighborhood of N: m1 = (1,0.0)
nbd of AMD: (0,0,1)
```
$W(m_1,m_2)=2\times 1$, move 2 steps with probability 1


**Ollivier–Ricci curvature compares how probability mass around neighboring nodes is transported.**

Edge Ricci curvature uses `average of Ricci curvature`,

Practical interpretation in this project:

- **negative curvature:** stretched or bridge-like relationship;
- **near zero:** locally neutral geometry;
- **positive curvature:** cohesive or redundant local relationship.

The implementation uses `GraphRicciCurvature` when available and a fallback approximation otherwise.

````python
G_ricci = compute_ricci_curvature(
    G_capital,
    alpha=0.5,
    method="OTD",
    proc=1,
)

 

edge_table = summarize_edges(G_ricci)
display(edge_table.sort_values("ricciCurvature"))
````

## 11. Ricci flow: geometric regularization, not forecasting

Ricci flow updates graph edge lengths according to curvature. In v12 it is used as a geometric experiment:

### Why use Ricci flow
Three kind Graphs made:

|Graph|	Represents|	Main question|
|---|---|---|
|$G_0$|Statistical similarity|	Which stocks move together?|
|$G_{capital}$|Statistical + liquidity similarity|	Which stocks move together with compatible capital scale?|
|$G_{flowed}$|Geometry after Ricci Flow|	What is the intrinsic market manifold after removing local geometric distortions?|

The original graph is only a snapshot. Ricci Flow continuously adjusts edge lengths according to local curvature.

$$d_{ij}^{(t+1)}=d_{ij}^{(t)}\left(1-\eta\kappa_{ij}^{(t)}\right).$$

```
Positive curvature contracts edges.
Negative curvature expands edges.
```
After many iterations, the graph reaches an approximately stable geometry.

It should be interpreted as **regularized geometry**, not the future market state.

````python
G_flowed, flow_history = run_ricci_flow(
    G_ricci,
    iterations=8,
    step_size=0.25,
    alpha=0.5,
    method="OTD",
    proc=1,
)

display(flow_history)
display(compare_before_after_flow(G_ricci, G_flowed).head(10))
plot_ricci_flow_history(flow_history).show()
````

## 12. Build rolling frames

A rolling frame repeats the entire graph construction over a moving time window.

Example with a 60-day window and 5-day step:

```text
Frame 1: days 1–60
Frame 2: days 6–65
Frame 3: days 11–70
```

Each frame stores the graph, correlations, distances, clusters, capital attributes, curvature, and graph-level statistics.

````python
frames, starts = build_rolling_frames(
    returns=returns,
    window_size=60,
    step=5,
    max_frames=30,
    graph_mode="knn+bridges",
    k_neighbors=3,
    min_corr=0.05,
    max_bridges=3,
    min_pair_obs=20,
    alpha=0.5,
    method="OTD",
    proc=1,
    dollar_volume=dollar_volume,
    use_capital_weighting=True,
    capital_alpha=0.35,
)

print("Frames:", len(frames))
print("First frame date:", frames[0].stats.end_date)
print("Last frame date:", frames[-1].stats.end_date)
````

## 14. Gaussian Hidden Markov Model

### Why use HMM
Looking only at returns, the market appears noisy, random. HMM can infer hidden states, bull or bear, neutral, which fluctuations are consistent with one underlying regime.

HMM separates:
- short-term observation noise
- longer-lasting hidden regimes


Formular
---
Let $S_t$ be a hidden market-structure state and \(Y_t\) be the observed graph-feature vector. The HMM models:


$$P(S_t\mid S_{t-1})$$

and

$$P(Y_t\mid S_t).$$

v12 standardizes the features, fits a diagonal-covariance Gaussian HMM, predicts the most likely state, and calculates posterior probabilities.

Now HMM estimates
```
the hidden geometric regime of the market
```
rather than only the return regime. Advagtage of using HMM:
|Feature|interretation|
|---|---|
|high average Ricci|Stable market.|
|negative Ricci|Market under structural stress|
|positive Ricci|Rotation regime|

Features of HMM_df
---
|Feature|	Meaning	| Why uses|
|---|---|---|
|mean_return|	Average return of all stocks in the window|	Market direction|
|volatility|	Standard deviation of returns|	Market risk/stress|
|avg_ricci|	Mean Ollivier/Forman Ricci curvature|	Overall network robustness|
|ricci_std|	Variation of Ricci curvature|Structural heterogeneity|
|capital_similarity|	Average $S_{ij}$|Liquidity consistency|
|flow_energy (or delta2)|	Distance between $G_{capital}$ and $G_{flowed}$|Geometric stress|
|num_clusters|	Number of connected/clustered communities|	Market fragmentation|
|largest_component_ratio|	Fraction of nodes in the largest component|	Connectivity|
|bridge_ratio|	Fraction of negative-curvature bridge edges|	Structural bottlenecks|

### Output

|Feature|	Meaning|
|---|---|
|state|	Most likely hidden regime (0, 1, 2, …)|
|state_probability| (optional)	Confidence of the inferred state|
|regime_name|	Human-readable label such as Stable, Rotation, Stress|

````python
hmm_df, regime_labels = compute_hmm_regimes(
    frames=frames,
    returns=returns,
    starts=starts,
    window_size=60,
    n_components=3,
    forward_days=5,
    random_state=42,
)

print("Regime labels:", regime_labels)
display(hmm_df.tail())
plot_hmm_regimes(hmm_df).show()
````

## 15. Posterior probabilities

A hard state label hides uncertainty. Posterior probabilities answer a richer question:

> How much probability does the model assign to each hidden regime at this frame?

A frame labeled “transition” with 52% confidence is qualitatively different from one labeled “transition” with 96% confidence.

````python
probability_columns = [
    column
    for column in hmm_df.columns
    if column.startswith("state_probability_")
]

if probability_columns:
    display(
        hmm_df[
            ["date", "hmm_state", "regime_name", "state_probability"]
            + probability_columns
        ].tail(10)
    )
else:
    print("Posterior columns are unavailable; check hmmlearn installation and frame count.")
````

## 16. Generate a frame-by-frame market story

For every frame \(t\), v12 compares graph statistics with frame \(t-1\):


$$\Delta X_t=X_t-X_{t-1}.$$

The story is generated only from measured thresholds. It can say that curvature weakened or fragmentation increased, but it does not claim that these changes caused a future price move.

````python
stories = build_frame_stories(frames)
story_df = frame_story_table(stories)

display(
    story_df[
        [
            "frame",
            "date",
            "regime_name",
            "state_probability",
            "headline",
            "strongest_fragile_edge",
            "strongest_capital_edge",
        ]
    ].tail(10)
)

print("Latest narrative:\n")
print(stories[-1].narrative)
````

## 19. Research interpretation and limitations

Before using the results, remember:

1. correlation is not causation;
2. edge capital flow is a proxy, not direct fund-flow measurement;
3. HMM state numbers and labels are sample-dependent;
4. Ricci curvature describes graph geometry, not expected return;
5. Ricci flow is regularization, not a temporal forecast;
6. automated stories depend on thresholds in `story.py`;
7. changing tickers, windows, and graph construction can change the regimes;
8. all claims should be evaluated with walk-forward validation.
