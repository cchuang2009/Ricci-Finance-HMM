## 4. Capital concentration features / 資金集中度特徵

For normalized capital shares $p_i$, concentration is the Herfindahl index:


$$H = \sum_i p_i^2.$$


Low values indicate dispersed capital; high values indicate concentration in a small number of nodes or edges.

````python
def concentration_index(values) -> float:
    x = np.asarray(list(values), dtype=float)
    x = x[np.isfinite(x)]
    x = np.clip(x, 0.0, None)
    total = float(x.sum())
    if x.size == 0 or total <= 0.0:
        return 0.0
    shares = x / total
    return float(np.sum(shares ** 2))


def capital_feature_row(graph: nx.Graph) -> dict[str, float]:
    node_values = [
        data.get("capital_mass", data.get("capital", 0.0))
        for _, data in graph.nodes(data=True)
    ]
    edge_values = [
        data.get("edge_capital_flow", data.get("capital_flow", 0.0))
        for _, _, data in graph.edges(data=True)
    ]
    return {
        "capital_concentration": concentration_index(node_values),
        "edge_capital_concentration": concentration_index(edge_values),
    }

print(capital_feature_row(nx.Graph()))
````

## 5. Jupyter-safe rolling Ricci network / Jupyter 安全的滑動 Ricci 網路

This orchestration function calls the lower-level `ricci_finance` functions directly. It deliberately avoids the legacy `build_frames()` and therefore does not create Streamlit `ScriptRunContext`

Instead of representing the entire historical period with one static graph, it repeatedly moves a fixed-length window through time:
```
Window 1 → Graph 1 → Ricci features 1
Window 2 → Graph 2 → Ricci features 2
...
```
For each window, it performs this pipeline
```
Returns within one window
        ↓
Correlation and distance matrices
        ↓
Market graph
        ↓
Capital attributes
        ↓
Ollivier–Ricci curvature
        ↓
Graph-level features
        ↓
Time-indexed market state
```
Main function, `build_rolling_ricci_frames(**parameters)`, converts a time series of market returns and trading capital into a sequence of rolling Ricci-curvature market graphs. 

### Returns
- frames: detailed graph snapshots used for visualization, surgery, sector analysis, and animation.
- feature_df: one numerical feature row per graph snapshot, used later by the HMM to identify market regimes.

### Parameters
-    window: number of observations in each rolling window.
-    step: number of rows by which the rolling window moves after each graph
-    max_frames: Limits the number of rolling frames
-    k: number of nearest or strongest neighbors used during graph construction.
```
A larger k generally creates:
  more edges; a denser graph; more connected sectors;
  greater computational cost.

A smaller k creates:
  a sparse graph; stronger emphasis on the closest relationships;
  a greater chance of disconnected components.
```
-    min_corr: an edge between stocks i and j is retained only when their relationship is sufficiently strong:
$$\rho_{ij}\ge\rho_{\text{min}}$$
    the more higher threshold produces a selective graph containing only sthe more trong relationships.
-    max_bridges: Controls how many additional bridge edges may be introduced to prevent the graph from becoming excessively fragmented.
-    min_obs: The minimum number of valid paired observations required to calculate a relationship between two assets; this is especially important when: a stock has a recent IPO, a ticker has missing data, etc.
-    alpha: The Ollivier–Ricci idleness parameter, within 0 and 1, which keeps the fact: in each discrete transport interpretation, part of the probability mass remains at the original node, while the rest is distributed among neighboring nodes.
-    method:exact accepted values depend on the installed ricci_finance and GraphRicciCurvature implementation.
-    proc: The number of worker processes used in Ricci-curvature computation.
-    capital_alpha: Controls how strongly dollar-volume information affects the capital attributes attached to nodes and edges.
    use_capital: Enables or disables capital-aware graph attributes.


````python
def build_rolling_ricci_frames(
    returns: pd.DataFrame,
    dollar_volume: pd.DataFrame,
    *,
    window: int,
    step: int,
    max_frames: int | None,
    k: int,
    min_corr: float,
    max_bridges: int,
    min_obs: int,
    alpha: float,
    method: str,
    proc: int,
    capital_alpha: float,
    use_capital: bool,
):
    starts = list(range(0, max(0, len(returns) - window + 1), step))
    if max_frames and len(starts) > max_frames:
        starts = starts[-max_frames:]

    frames = []
    rows = []
    previous_graph = None

    for number, start in enumerate(starts, start=1):
        end = start + window
        window_returns = returns.iloc[start:end]
        window_dollar_volume = dollar_volume.reindex(window_returns.index)

        graph, correlation, distance = build_graph_from_window(
            window_returns,
            k,
            min_corr,
            max_bridges,
            min_obs,
        )
        graph = attach_capital_attributes(
            graph,
            window_dollar_volume,
            capital_alpha,
            use_capital,
        )
        graph = compute_ricci_curvature(
            graph,
            alpha,
            method,
            proc,
        )

        date = window_returns.index[-1]
        row = graph_feature_row(graph, date)
        row.update(capital_feature_row(graph))
        row["edge_stability"] = (
            edge_jaccard(previous_graph, graph)
            if previous_graph is not None
            else 1.0
        )

        frames.append({
            "graph": graph,
            "date": date,
            "correlation": correlation,
            "distance": distance,
        })
        rows.append(row)
        previous_graph = graph

        print(
            f"\rFrame {number}/{len(starts)} | {pd.Timestamp(date).date()} | "
            f"nodes={graph.number_of_nodes()} edges={graph.number_of_edges()}",
            end="",
        )

    print()
    feature_df = pd.DataFrame(rows)
    if not feature_df.empty:
        feature_df["date"] = pd.to_datetime(feature_df["date"])
    return frames, feature_df
````

## 6. Build rolling frames / 建立滑動網路

````python
frames, feature_df = build_rolling_ricci_frames(
    returns=returns,
    dollar_volume=dollar_volume,
    window=WINDOW,
    step=STEP,
    max_frames=MAX_FRAMES,
    k=K_NEIGHBORS,
    min_corr=MIN_CORR,
    max_bridges=MAX_BRIDGES,
    min_obs=MIN_PAIR_OBS,
    alpha=RICCI_ALPHA,
    method=RICCI_METHOD,
    proc=RICCI_PROC,
    capital_alpha=CAPITAL_ALPHA,
    use_capital=USE_CAPITAL,
)

if len(feature_df) < max(12, N_STATES * 4):
    raise ValueError(
        f"Only {len(feature_df)} frames were built. Increase PERIOD or "
        "reduce WINDOW/STEP."
    )

print("Frames:", len(frames))
#print("Feature columns:\n", feature_df.columns.tolist())
display(feature_df.tail(2).T)
````

## 7. Validate HMM features / 驗證 HMM 特徵

This cell prevents the pandas error:

`['capital_concentration', 'edge_capital_concentration'] not in index`

````python
def validate_hmm_features(
    frame: pd.DataFrame,
    requested: list[str],
) -> list[str]:
    missing = [name for name in requested if name not in frame.columns]
    if missing:
        raise ValueError(
            "Required HMM features were not generated: "
            f"{missing}. Available columns: {frame.columns.tolist()}"
        )
    return list(requested)

HMM_FEATURES = validate_hmm_features(
    feature_df,
    list(DEFAULT_HMM_FEATURES),
)

print("Validated HMM features:")
#print(HMM_FEATURES)
display(feature_df[HMM_FEATURES].describe().T)
````

## 11. One-week regime forecast / 一週狀態預測

The five-trading-day forecast compares transition-matrix propagation, current-state propagation, Monte Carlo simulation and duration-aware adjustment where supported by the installed V15 module.

````python
# Number of HMM regimes
n_states = hmm_model.n_components

# Default regime labels
regime_labels = {
    i: f"Regime {i}"
    for i in range(n_states)
}

print(regime_labels)
````

````python
FORECAST_HORIZON = 5

forecast_df = forecast_hmm_methods(
    hmm_model,
    posterior_last=np.asarray(posterior[-1], dtype=float),
    viterbi_last=int(hmm_states[-1]),
    horizon=FORECAST_HORIZON,
    n_sim=5000,
    random_state=RANDOM_STATE,
    current_run_length=current_run_length(hmm_states),
)

forecast_df["regime"] = forecast_df["state"].map(regime_labels)

forecast_table = forecast_df.pivot(
    index="regime",
    columns="method",
    values="probability",
)

display((forecast_table * 100).round(2).astype(str) + "%")

print("Current Viterbi state:", int(hmm_states[-1]))
print("Current run length:", current_run_length(hmm_states))
print("Historical switch rate:", switch_rate(hmm_states))
````

````python

## 16. Optional pure-PyTorch GNN comparison / 可選純 PyTorch GNN 比較

The GNN section is optional and does **not** require PyTorch Geometric.

Its purpose is to test whether rolling Ricci-network snapshots contain enough information to reproduce, or optionally forecast, HMM regimes.

### Interpretation limits

1. The labels are produced by the HMM, so this is not an independent ground truth.
2. A small number of rolling frames can make accuracy unstable.
3. Rare HMM states may be absent from the chronological training period.
4. Same-date prediction measures imitation; next-state prediction is more useful but harder.
5. Balanced accuracy is important when regime classes are imbalanced.
````

````python

import inspect
import numpy as np
import pandas as pd

all_nodes = sorted({
    node
    for frame in frames
    for node in frame["graph"].nodes()
})

all_sectors = assign_sectors(all_nodes)
graph_list = [frame["graph"] for frame in frames]
gnn_labels = np.asarray(hmm_states, dtype=int)

n_samples = min(len(graph_list), len(gnn_labels), len(feature_df))
if n_samples < 8:
    raise ValueError(
        f"Only {n_samples} graph/label samples are available. "
        "Increase PERIOD or reduce WINDOW/STEP before testing the GNN."
    )

graph_list = graph_list[:n_samples]
gnn_labels = gnn_labels[:n_samples]
gnn_dates = pd.to_datetime(feature_df["date"].iloc[:n_samples]).to_numpy()

print("GNN samples:", n_samples)
print("Class counts:", dict(zip(*np.unique(gnn_labels, return_counts=True))))

# Pass only arguments supported by the installed ricci_finance revision.
candidate_kwargs = {
    "epochs": 250,
    "hidden": 32,
    "dropout": 0.25,
    "learning_rate": 0.003,
    "weight_decay": 1e-4,
    "train_fraction": 0.70,
    "patience": 40,
    "random_state": RANDOM_STATE,
    "use_cuda": True,
    "predict_next_state": False,
}

signature = inspect.signature(train_gcn_regime)
supported_kwargs = {
    key: value
    for key, value in candidate_kwargs.items()
    if key in signature.parameters
}

try:
    gnn_result = train_gcn_regime(
        graph_list,
        gnn_labels,
        all_sectors,
        **supported_kwargs,
    )

    print(getattr(gnn_result, "note", "GNN completed."))
    print("Device:", getattr(gnn_result, "device", "not reported"))
    print("Epochs:", getattr(gnn_result, "epochs", "not reported"))
    print("Accuracy:", getattr(gnn_result, "accuracy", np.nan))
    print(
        "Balanced accuracy:",
        getattr(gnn_result, "balanced_accuracy", "not reported"),
    )

except ImportError as exc:
    gnn_result = None
    print("PyTorch is not installed; GNN section skipped:", exc)

except Exception as exc:
    gnn_result = None
    print("GNN section failed:", type(exc).__name__, exc)
````

````python

comparison = pd.DataFrame()

if gnn_result is not None:
    predictions = np.asarray(
        getattr(gnn_result, "predictions", []),
        dtype=int,
    )

    if predictions.size:
        result_labels = np.asarray(
            getattr(gnn_result, "labels", gnn_labels),
            dtype=int,
        )

        test_indices = getattr(gnn_result, "test_indices", None)

        if test_indices is not None:
            test_indices = np.asarray(test_indices, dtype=int)
            comparison = pd.DataFrame({
                "date": gnn_dates[test_indices],
                "HMM": result_labels[test_indices],
                "GCN": predictions[test_indices],
            })

            probabilities = getattr(gnn_result, "probabilities", None)
            if probabilities is not None:
                probabilities = np.asarray(probabilities)
                for class_index in range(probabilities.shape[1]):
                    comparison[f"P_class_{class_index}"] = (
                        probabilities[test_indices, class_index]
                    )
        else:
            # Compatibility with older revisions returning test predictions only.
            n_pred = len(predictions)
            comparison = pd.DataFrame({
                "date": gnn_dates[-n_pred:],
                "HMM": result_labels[-n_pred:],
                "GCN": predictions,
            })

        display(comparison)

        print("\nConfusion matrix")
        display(pd.crosstab(
            comparison["HMM"],
            comparison["GCN"],
            rownames=["Actual HMM"],
            colnames=["Predicted GCN"],
            dropna=False,
        ))
    else:
        print("The GNN completed but returned no predictions.")
else:
    print("No GNN result is available.")
````

# Interpretation checklist / 結果解讀清單

1. **Average Ricci curvature:** lower or more negative values indicate a more fragile network geometry.
2. **Negative-edge ratio:** the proportion of potentially stressed relationships.
3. **Edge stability:** persistence of network connections between rolling windows.
4. **Capital concentration:** whether node capital is broadly distributed or dominated by a few tickers.
5. **Edge-capital concentration:** whether estimated capital flow is spread across many edges or concentrated in a few channels.
6. **HMM state probability:** confidence in the decoded state, not certainty about future returns.
7. **Sector flow:** estimated structural flow based on graph and dollar-volume features; it is not audited transaction-level fund flow.
